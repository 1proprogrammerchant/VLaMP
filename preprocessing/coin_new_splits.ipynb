```python
import json
```

```python
with open('../.data/coin/COIN.json') as f:
    coin_all = json.load(f)
```

```python
with open('../.data/coin/train_ours.txt') as f:
    train_ids = f.readlines()
```

```python
with open('../.data/coin/validation_ours.txt') as f:
    val_ids = f.readlines()
```

```python
with open('../.data/coin/test_ours.txt') as f:
    test_ids = f.readlines()
```

```python

all_valid = set(train_ids + val_ids + test_ids)
```

```python
new_train = []
new_val = []
new_test = []
skipped = 0
for idx, v in coin_all["database"].items():
    subset  = v['subset']
    if idx not in all_valid:
        skipped+=1
        print(f"skipped {idx}. Total {skipped}")
        continue
    if subset == "training":
        new_train.append(idx)
    elif subset == "testing":
        new_test.append(idx)
    else:
        raise ValueError(f"subset {subset}")


```

```python
def strip(l):
    return [i.strip() for i in l]
```

```python
train_ids, val_ids, test_ids = set(strip(train_ids)), set(strip(val_ids)), set(strip(test_ids))
```

```python
missing = 0
for inp_idx, v in coin_all['database'].items():
    if inp_idx in train_ids:
        v['subset'] = "training"
    elif inp_idx in val_ids:
        v['subset'] = "training"
    elif inp_idx in test_ids:
        v["subset"] = "testing"
    else:
        missing +=1
```

```python
with open('.data/coin/COIN_new_split.json', 'w') as f:
    json.dump(coin_all,f, indent=4)
```
